```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def f(z, n=100):
    f = 0
    for i in range(n):
        f = f**2 + z
    return f
```

```python
num = 300
grid_x = np.linspace(-1.7, 0.6, num=num)
grid_y = np.linspace(-1.4, 1.4, num=num)
mandelbrot_image = np.zeros((grid_x.size, grid_y.size), dtype=np.complex64)
```

```python
for ix, x in enumerate(grid_x):
    for iy, y in enumerate(grid_y):
        mandelbrot_image[ix, iy] = f(x + 1j * y)
```

```python
extent = (min(grid_x), max(grid_x), min(grid_y), max(grid_y))
plt.imshow(abs(mandelbrot_image.T)**2 < 1000, cmap = cm.Reds,  extent=extent)
plt.show()
```
