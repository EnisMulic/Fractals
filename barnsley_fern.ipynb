```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def generate_stem(x, y):
    return (0.0, 0.16 * y)
def generate_successively_smaller_leaflets(x, y):
    return (0.85 * x + 0.04 * y, -0.04 * x + 0.85 * y + 1.60)
def generate_largest_left_hand_leaflet(x, y):
    return (0.20 * x - 0.26 * y, 0.23 * x + 0.22 * y + 1.60)
def generate_largest_right_hand_leaflet(x, y):
    return (-0.15 * x + 0.28 * y, 0.26 * x + 0.24 * y + 0.44)
transformations = [generate_stem, generate_successively_smaller_leaflets, generate_largest_left_hand_leaflet, generate_largest_right_hand_leaflet]
```

```python
width, height = 300, 300
fern_image = np.zeros((width, height))
points = 100000
x, y = 0, 0
for i in range (points):
    transform = np.random.choice(transformations, p = [0.01, 0.85, 0.07, 0.07])
    x, y = transform(x, y)
    ix, iy = int(width / 2 + x * width / 10), int(y * height / 12)
    fern_image[iy, ix] = 1
```

```python
plt.imshow(fern_image[::-1,:], cmap = cm.Greens)
plt.axis('off')
plt.show()
```
